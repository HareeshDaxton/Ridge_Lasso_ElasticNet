```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, 'Region']= 0
df.loc[122:, 'Region']= 1
```

```python
df[['Region']] = df[['Region']].astype('int64')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.loc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.loc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype('int64')
```

```python
df.info()
```

```python
objects = [fet for fet in df.columns if df[fet].dtype == 'object']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df.to_csv('Cleaned_Algerian_forest_fires_dataset.csv', index=False)
```
